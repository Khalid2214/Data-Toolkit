Ans1: NumPy (Numerical Python) is a powerful library for numerical computing in Python. It provides:

Multidimensional arrays (ndarray) that are faster and more memory-efficient than Python lists.

Mathematical functions to operate on arrays element-wise.

Linear algebra, Fourier transforms, random number generation, and other advanced capabilities.

Why it's widely used:

Efficient storage and manipulation of numerical data.

Foundation for other libraries like Pandas, SciPy, Scikit-learn, and TensorFlow.

Supports vectorized operations (i.e., avoids explicit loops, making code faster and cleaner).

Ans2: Broadcasting is a technique that allows NumPy to perform arithmetic operations on arrays of different shapes.

Rules:

If the arrays have different numbers of dimensions, the smaller one is padded with ones on the left.

Dimensions are compatible when:

They are equal, or

One of them is 1.

```python
#Ans2

import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'Score': [85, 90, 78]
}

df = pd.DataFrame(data)

# Calculate mean of the 'Score' column
mean_score = df['Score'].mean()
print("Mean Score:", mean_score)
```

```python
#Ans3

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 7, 8, 5, 9]

# Scatter plot
plt.scatter(x, y, color='blue', marker='o')
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
#Ans4

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [85, 90, 78, 92],
    'Science': [88, 85, 82, 95],
    'English': [90, 89, 85, 87]
}
df = pd.DataFrame(data)

# Correlation matrix
corr = df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Ans6

import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [80, 90, 85]
})

# Add a new column: Grade based on Score
df['Grade'] = df['Score'].apply(lambda x: 'A' if x >= 90 else 'B' if x >= 80 else 'C')

print(df)
```

```python
#Ans11

import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication
C = np.matmul(A, B)
# OR using the @ operator: C = A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", C)
```

```python
#Ans13

import plotly.express as px
import pandas as pd
import numpy as np

# Sample 3D data
df = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'z': np.random.rand(50),
    'category': np.random.choice(['A', 'B'], 50)
})

# 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='category',
                    title="3D Scatter Plot with Plotly")
fig.show()
```
